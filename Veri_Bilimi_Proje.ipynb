```python
# Takım başına sporcu sayısını hesapla
team_counts = Athlets['Team'].value_counts().reset_index()
team_counts.columns = ['Team', 'Count']  # Kolon adlarını düzenle

# 100'den fazla sporcusu olan takımları filtrele
filtered_team_counts = team_counts[team_counts['Count'] > 100]

# Çubuk grafiği plotly ile oluşturma ve çubukların içine sayıları yazma
fig = px.bar(
    filtered_team_counts,
    x='Team',
    y='Count',
    text='Count',  # Çubukların içine sporcu sayısını yerleştir
    labels={'Team': 'Takımlar', 'Count': 'Sporcu Sayısı'},
    title="Sporcu Sayısı 100'den Fazla Olan Takımlar"
)

# Metinleri çubukların içine yerleştirme ayarları
fig.update_traces(texttemplate='%{text}', textposition='inside')

# Grafik gösterimi
fig.show()
```

```python
Athlets.groupby("Team")["Discipline"].value_counts()
```

```python
# 'Team' ve 'Discipline' sütunlarına göre gruplama ve sayma
grouped = Athlets.groupby(['Team', 'Discipline']).size().reset_index(name='Counts')

# Takım başına toplam sporcu sayısını bulma
team_counts = grouped.groupby('Team')['Counts'].sum()

# 100'den fazla sporcusu olan takımları filtrele
teams_with_more_than_100 = team_counts[team_counts > 100].index

# Bu takımları orijinal grup verisi ile eşleştir
filtered_grouped = grouped[grouped['Team'].isin(teams_with_more_than_100)]

# Interaktif yığılmış çubuk grafiği plotly ile oluşturma
fig = px.bar(filtered_grouped,
             x='Team',
             y='Counts',
             color='Discipline',
             title="100'den Fazla Sporcusu Olan Takımların Sporcu Dağılımı",
             labels={'Counts': 'Sporcu Sayısı', 'Team': 'Takımlar', 'Discipline': 'Disiplinler'},
             hover_data={'Discipline': True, 'Counts': True})

# Grafik gösterimi
fig.show()
```

```python
medal_counts = [Medals['gold'].sum(),Medals['silver'].sum(),Medals['bronze'].sum()]
medal_labels = ['Gold','Silver','Bronze']
fig = px.pie(values=medal_counts,names=medal_labels,title='Distribution by Olympics 2024')
fig.show()
```

```python
# Veriyi melt ile uzatma
medals_long = Medals.melt(id_vars='Nationality', value_vars=['gold', 'silver', 'bronze'],
                              var_name='Medal', value_name='Count')

# Yığılmış çubuk grafiği oluşturma
fig = px.bar(
    medals_long,
    x='Nationality',
    y='Count',  # Y ekseninde madalya sayıları
    color='Medal',  # Renk için madalya türü
    title="Madalya Sayıları Ülkelere Göre",
    text='Count',  # Çubukların içine madalya sayısını yazdır
    category_orders={'Nationality': Medals['Nationality']}  # Tüm ülkeleri göstermek için
)

# Metinleri çubukların içine yerleştirme ayarları
fig.update_traces(texttemplate='%{text}', textposition='inside')

# Grafik gösterimi
fig.show()
```

```python
Medals[Medals["Nationality"] == "Türkiye"]
```

```python
# En fazla madalya kazanan 20 ülkeyi seçme
top_20_countries = Medals.nlargest(20, 'total')

# Veriyi melt ile uzatma
medals_long = top_20_countries.melt(id_vars='Nationality', value_vars=['gold', 'silver', 'bronze'],
                                      var_name='Medal', value_name='Count')

# Yığılmış çubuk grafiği oluşturma
fig = px.bar(
    medals_long,
    x='Nationality',
    y='Count',  # Y ekseninde madalya sayıları
    color='Medal',  # Renk için madalya türü
    title="En Fazla Madalya Kazanan 20 Ülke",
    text='Count',  # Çubukların içine madalya sayısını yazdır
)

# Metinleri çubukların içine yerleştirme ayarları
fig.update_traces(texttemplate='%{text}', textposition='inside')

# Grafik gösterimi
fig.show()
```
